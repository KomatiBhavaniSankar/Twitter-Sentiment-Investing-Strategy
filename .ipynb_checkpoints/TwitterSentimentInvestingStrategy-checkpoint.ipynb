```python
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import yfinance as yf
import os

plt.style.use('ggplot')
sentiment_df = pd.read_csv('sentiment_data.csv')
sentiment_df['date'] =pd.to_datetime(sentiment_df['date'])
sentiment_df = sentiment_df.set_index(['date','symbol'])
sentiment_df['engagement_ratio'] = sentiment_df['twitterComments']/sentiment_df['twitterLikes']
sentiment_df = sentiment_df[(sentiment_df['twitterLikes'] > 20) & (sentiment_df['twitterComments'] > 10)]
```

```python
aggragated_df =(sentiment_df.reset_index('symbol').groupby([pd.Grouper(freq='ME'),'symbol'])[['engagement_ratio']].mean())
aggragated_df['rank'] = aggragated_df.groupby(level=0)['engagement_ratio'].transform(lambda x:x.rank(ascending=False))                                                            
```

```python
aggragated_df
```

```python
filterd_df = aggragated_df[aggragated_df['rank']<6].copy()
filterd_df = filterd_df.reset_index(level=1)
filterd_df.index = filterd_df.index+pd.DateOffset(1)
filterd_df = filterd_df.reset_index().set_index(['date','symbol'])
```

```python
filterd_df
```

```python
dates = filterd_df.index.get_level_values('date').unique().tolist()
fixed_dates = {}
for d in dates:
    fixed_dates[d.strftime('%Y-%m-%d')] = filterd_df.xs(d, level=0).index.tolist()
fixed_dates
```
